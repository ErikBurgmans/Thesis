```python
def calculate_euclidean_mouth_eyes(csv_files): #this creates the lips and eyes extra features dataframe
    df_output =pd.DataFrame(columns=['top_lip_outer_max', 'top_lip_outer_mean', 'top_lip_outer_min', 'top_lip_inner_max',
                                'top_lip_inner_mean', 'top_lip_inner_min', 'corner_lip_outer_max', 'corner_lip_outer_mean',
                                'corner_lip_outer_min', 'corner_lip_inner_max', 'corner_lip_inner_mean', 'corner_lip_inner_min',
                                'left_eye_left_max', 'left_eye_left_mean', 'left_eye_left_min', 'left_eye_right_max',
                                'left_eye_right_mean', 'left_eye_right_min', 'left_eye_corner_max', 'left_eye_corner_mean',
                                'left_eye_corner_min', 'right_eye_left_max', 'right_eye_left_mean', 'right_eye_left_min',
                                'right_eye_right_max', 'right_eye_right_mean', 'right_eye_right_min', 'right_eye_corner_max',
                                'right_eye_corner_mean', 'right_eye_corner_min'])
    
    list_lips1 = [350, 361, 347, 359] #order: top lip outer, top lip inner, left cornor outer, left cornor inner
    list_lips2 = [356, 365, 353, 363]  #order: bottom lip outer, bottom lip inner, right cornor outer, right cornor inner
    
    list_l_eye1 = [336, 338, 335] #order: top left, top right, left cornor
    list_l_eye2 = [340, 339, 338] #order: bottom left, bottom right, right cornor
    
    list_r_eye1 = [342, 343, 341] #order: top left, top right, left cornor
    list_r_eye2 = [346, 345, 344] #order: bottom left, bottom right, right cornor

    def calculate_euclidean (x1, x2, y1, y2):
        output = math.sqrt(((x2-x1)**2)+((y2-y1)**2))
        return output

    def calculate_euclidean_lips(list_lips1, list_lips2, df):
        outputlist = []
        for i in range(0, len(list_lips1)):
            list1 = []
            for j in range(0, len(df.iloc[:,list_lips1[i]])):
                x = calculate_euclidean(df.iloc[j,list_lips1[i]], df.iloc[j,list_lips2[i]], df.iloc[j,list_lips1[i]+68], df.iloc[j,list_lips2[i]+68])
                list1 = np.append(list1, x)
            y = np.max(list1)
            z = np.mean(list1)
            a = np.min(list1)

            outputlist = np.append(outputlist, y)
            outputlist = np.append(outputlist, z)
            outputlist = np.append(outputlist, a)
        return outputlist

    def calculate_euclidean_eye_left(list_l_eye1, list_l_eye2, df):
        outputlist = []
        for i in range(0, len(list_l_eye1)):
            list1 = []
            for j in range(0, len(df.iloc[:,list_l_eye1[i]])):
                x = calculate_euclidean(df.iloc[j,list_l_eye1[i]], df.iloc[j,list_l_eye2[i]], df.iloc[j,list_l_eye1[i]+68], df.iloc[j,list_l_eye2[i]+68])
                list1 = np.append(list1, x)
            y = np.max(list1)
            z = np.mean(list1)
            a = np.min(list1)

            outputlist = np.append(outputlist, y)
            outputlist = np.append(outputlist, z)
            outputlist = np.append(outputlist, a)
        return outputlist


    def calculate_euclidean_eye_right(list_r_eye1, list_r_eye2, df):
        outputlist = []
        for i in range(0, len(list_r_eye1)):
            list1 = []
            for j in range(0, len(df.iloc[:,list_r_eye1[i]])):
                x = calculate_euclidean(df.iloc[j,list_r_eye1[i]], df.iloc[j,list_r_eye2[i]], df.iloc[j,list_r_eye1[i]+68], df.iloc[j,list_r_eye2[i]+68])
                list1 = np.append(list1, x)
            y = np.max(list1)
            z = np.mean(list1)
            a = np.min(list1)

            outputlist = np.append(outputlist, y)
            outputlist = np.append(outputlist, z)
            outputlist = np.append(outputlist, a)
        return outputlist
    
    count = 1
    
    
    for f in csv_files:
        outputlist = []
        print(count,'/ 6000')
        df=pd.read_csv(f)
        mouth = calculate_euclidean_lips(list_lips1, list_lips2, df)
        left_eye = calculate_euclidean_eye_left(list_l_eye1, list_l_eye2, df)
        right_eye = calculate_euclidean_eye_right(list_r_eye1, list_r_eye2, df)
        outputlist = np.append(outputlist, mouth)
        outputlist = np.append(outputlist, left_eye)
        outputlist = np.append(outputlist, right_eye)
        df_output = df_output.append(pd.DataFrame(outputlist.reshape(1, -1), columns=list(df_output)), ignore_index=True)
        count += 1
    
    return df_output
```

```python
def labels_dataframe (dictionary, df): #this labels all the dataframe
    
    def label_openness(dictionary, df):
        df_openness = pd.DataFrame(columns=['Openness'])

        count = 0
        for i in range(0, len(df)):
            append_dataframe1 = pd.DataFrame([dictionary['openness'][df.iloc[i,0]]], columns = ['Openness'])
            df_openness = df_openness.append(append_dataframe1, ignore_index=True)
            count += 1
            print('1/6 |', count, '/', len(df), 'done')
        return df_openness
    
    def label_conscientiousness(dictionary, df):
        df_conscientiousness = pd.DataFrame(columns=['Conscientiousness'])

        count = 0
        for i in range(0, len(df)):
            append_dataframe2 = pd.DataFrame([dictionary['conscientiousness'][df.iloc[i,0]]], columns = ['Conscientiousness'])
            df_conscientiousness = df_conscientiousness.append(append_dataframe2, ignore_index=True)
            count += 1
            print('2/6 |', count, '/', len(df), 'done')
        return df_conscientiousness
    
    def label_extraversion(dictionary, df):
        df_extraversion = pd.DataFrame(columns=['Extraversion'])

        count = 0
        for i in range(0, len(df)):
            append_dataframe3 = pd.DataFrame([dictionary['extraversion'][df.iloc[i,0]]], columns = ['Extraversion'])
            df_extraversion = df_extraversion.append(append_dataframe3, ignore_index=True)
            count += 1
            print('3/6 |', count, '/', len(df), 'done')
        return df_extraversion

    def label_agreeableness(dictionary, df):
        df_agreeableness = pd.DataFrame(columns=['Agreeableness'])

        count = 0
        for i in range(0, len(df)):
            append_dataframe4 = pd.DataFrame([dictionary['agreeableness'][df.iloc[i,0]]], columns = ['Agreeableness'])
            df_agreeableness = df_agreeableness.append(append_dataframe4, ignore_index=True)
            count += 1
            print('4/6 |', count, '/', len(df), 'done')
        return df_agreeableness
    
    def label_neuroticism(dictionary, df):
        df_neuroticism = pd.DataFrame(columns=['Neuroticism'])

        count = 0
        for i in range(0, len(df)):
            append_dataframe5 = pd.DataFrame([dictionary['neuroticism'][df.iloc[i,0]]], columns = ['Neuroticism'])
            df_neuroticism = df_neuroticism.append(append_dataframe5, ignore_index=True)
            count += 1
            print('5/6 |', count, '/', len(df), 'done')
        return df_neuroticism
    
    def label_interview(dictionary, df):
        df_interview = pd.DataFrame(columns=['Interview'])

        count = 0
        for i in range(0, len(df)):
            append_dataframe6 = pd.DataFrame([dictionary['interview'][df.iloc[i,0]]], columns = ['Interview'])
            df_interview = df_interview.append(append_dataframe6, ignore_index=True)
            count += 1
            print('6/6 |', count, '/', len(df), 'done')
        return df_interview
    
    df2 = label_openness(dictionary, df)
    df3 = label_conscientiousness(dictionary, df)
    df4 = label_extraversion(dictionary, df)
    df5 = label_agreeableness(dictionary, df)
    df6 = label_neuroticism(dictionary, df)
    df7 = label_interview(dictionary, df)
    
    final_output = pd.concat([df,df2, df3, df4, df5, df6, df7], axis=1)
    return final_output
```
